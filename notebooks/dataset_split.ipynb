# Split the data!

We split the data into train, test, and validation csv files on a per video level.
Since ADL videos are much longer (and thus produce more samples in our method), we
want to keep the proportion of ADL videos vs. Fall videos similar among the different
splits. Hence, we stratify by the `category` column.


```python
SPLIT = (0.7, 0.2, 0.1)

train_df, test_df, val_df = split_dataset(
    annotations=raw_dataset,
    output_path="data/Fall_Simulation_Data/",
    split=SPLIT,
    random_seed=42,
    stratify="category",
    save_splits=True,
)
```

```python
display("------------------- Train -------------------")
display(train_df.head())
display(train_df.describe())
display(train_df.info())
display(train_df["category"].value_counts() / len(train_df))

display("------------------- Test -------------------")
display(test_df.head())
display(test_df.describe())
display(test_df.info())
display(test_df["category"].value_counts() / len(test_df))

display("------------------- Val -------------------")
display(val_df.head())
display(val_df.describe())
display(val_df.info())
display(val_df["category"].value_counts() / len(val_df))
```
