```python
from pathlib import Path
import ffmpeg
import argparse
import logging
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sampler import Sampler
```

```python
sampler = Sampler("nothingfornow", "../data/Fall_Simulation_Data/annotations.csv")



# Generate Gaussian samples
sample_out = "../data/Fall_Simulation_Data/gaussian_samples/"
sampler.gaussianSampling(sample_out, samples_per_min=5)

# Generate stratified samples
sample_out = "../data/Fall_Simulation_Data/stratified_samples/"
sampler.stratifiedSampling(sample_out, samples_per_min=5)

# Generate uniform samples
sample_out = "../data/Fall_Simulation_Data/uniform_samples/"
sampler.uniformSampling(sample_out, samples_per_min=5)
```

```python
strat_cuts = [11.205, 15.037, 109.087, 90.509, 25.083, 25.39, 24.436, 41.434, 49.488, 42.348, 59.942, 58.854, 49.491]
gauss_cuts = [18.566, 23.691, 27.969, 32.616, 37.74, 40.564, 39.026, 16.879, 23.525, 26.907]
uni_cuts = [92.05, 52.303, 9.154, 55.435, 82.237, 52.079, 80.707, 61.084, 26.861, 24.157]


cuts_A = pd.DataFrame({'Time':strat_cuts,'Sampling technique':'Stratified'})
cuts_B = pd.DataFrame({'Time':gauss_cuts,'Sampling technique':'Gaussian'})
cuts_C = pd.DataFrame({'Time':uni_cuts,'Sampling technique':'Uniform'})

data = pd.concat([cuts_A,cuts_B,cuts_C])

sns.set()

sns.boxplot(data = data ,y = 'Sampling technique' ,x = 'Time')
plt.title('Sample times for demo video')
plt.show()
```
